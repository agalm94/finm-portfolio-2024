### 1.

Consider a Linear Factor Pricing Model (LFPM).

Which metric do we examine to understand its fit, (or errors)...
* given the estimated **time-series (TS)** test?
* given the estimated **cross-sectional (CS)** test?

<span style="color:red;">

* Time Series: We would look at Mean-Absolute-Error (MAE) to understand its fit. In this regression, we would measure MAE using the $\alpha$'s in the regression. Consequently, the formula would look something like
$$\frac{1}{n}\sum_{i=1}^n |\alpha_i|$$

* Cross-Sectional: We would again look at the MAE, but the MAE for this regression would be measured using the $\nu$. Consequently, the formula would look something like
$$\frac{1}{n}\sum_{i=1}^n |\nu_i|$$

And note that the $\nu$ comes from 
$$\hat{y} = \eta_i + \lambda_i \beta_i + \nu_i$$

</span>

### 2.

Consider the Arbitrage Pricing Theory (APT). Is it fair to say that it is more likely to work for sets of assets with low cross-correlation? Why or why not?

<span style="color:red;">

No, it is not fair to say this. APT forms no opinions about performance of an LFPM based on assets with low cross-correlation. Instead, it tells us that it is more likely to work when the errors of the regressions that we run are uncorrelated with one another. That is, we need regressions 
$$\tilde{r_i} = \alpha_i + \vec{\beta} \mathbf{x} + \epsilon_i$$

Where

$$corr(\epsilon_i, \epsilon_j) = 0 \qquad \textit{for $i \neq j$}$$

Under these conditions, APT says that we can (assuming no arbitrage in the market) perfectly any asset using the assets that comprise the factors in the first model above.

</span>

### 4.

Over longer horizons, do investments have higher Sharpe ratios? How is this issue relevant to long-term asset allocators such as Barnstable?

<span style="color:red;">

Assuming that mean returns are positive, yes--investments have higher Sharpe ratios over longer horizons. This is because we have that the Sharpe ratio is given by

$$SR = \frac{\mu}{\sigma}$$

Here, we will say that $\mu$ represents an asset's mean **log** returns and $\sigma$ represents an asset's **log** return risk. We know from lectures that $\mu$ will scale linear with a longer time horizon and **\sigma** will scale sublinearly (square root of time horizon). Consequently, as the time period increases, the $SR$ increases by $\sqrt{h}$, where $h$ represents the number of periods.

</span>

### 6.

Suppose investors are **not** mean-variance investors. If we find an investment with a Sharpe ratio higher than the "market", would this would be inconsistent with the CAPM?

<span style="color:red;">

First off, investors not optimizing the MV tradeoff contradicts one of the foundational assumptions of CAPM. Now, with that in mind, we also are able to write the CAPM cross-section regression as

$$SR_i = (\rho_{i, mkt}) SR_{mkt}$$

It is a fact that $-1 \leq \rho \leq 1$. Consequently, the Sharpe ratio of the market is the maximum Sharpe ratio under CAPM. Consequently, if we found an investment with a Sharpe ratio higher than the market, this would also be inconsistent with CAPM. 

</span>

To mitigate this ill-conditioned Covariance Matrix problem, we discussed regularizing the covariance matrix. To regularize the covariance matrix, we simply shrink the off-diagonal elements of the matrix, which has the effect of improving the conditioning number and making the inversion of the matrix in solution to the MVO optimization more stable.

```python
# NOTE: This code is very similar to the Fama-MacBeth model for time-varying Betas. Consider generalizing to handle that task
def OOS_strat(df, factors, start):
    y = df
    X = sm.add_constant(factors)

    forecast_err, null_err,oos_predictions,null_predictions = [], [],[],[]

    for i,j in enumerate(df.index):
        if i >= start:
            currX = X.iloc[:i]
            currY = y.iloc[:i]
            reg = sm.OLS(currY, currX, missing = 'drop').fit()
            null_forecast = currY.mean()
            reg_predict = reg.predict(X.iloc[[i]])
            actual = y.iloc[[i]]
            oos_predictions.append(reg_predict.T)
            null_predictions.append(pd.DataFrame([[reg_predict.index[0]]], columns = ['date'], index = [null_forecast]))
            forecast_err.append(reg_predict.values - actual)
            null_err.append(null_forecast.values - actual)
            
    RSS = (np.array(forecast_err)**2).sum()
    TSS = (np.array(null_err)**2).sum()
    predictions_df = pd.concat(oos_predictions).T.drop_duplicates()
    null_predictions_df = pd.concat(null_predictions).T
    
    return ((1 - RSS/TSS),reg,predictions_df,null_predictions_df)
```
